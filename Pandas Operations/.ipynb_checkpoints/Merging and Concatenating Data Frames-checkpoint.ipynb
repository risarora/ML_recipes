```python
#Import the pandas library
import pandas as pd
```

```python
# Create the data Frames
student_data = {'Name':['Tom Dick', 'Jack Welch ', 'Steven Smith', 'Ricky Sharma'],
        'Age':[18,17,19,20],
       'Course_Id':[1001,1002,1003,1001]
       }
Student1 = pd.DataFrame(student_data)

student_data2 = {'Name':['Ron Weasly', 'Hermione Granger ', 'Harry Potter', 'Dumbledore'],
        'Age':[18,17,19,20],
       'Course_Id':[1002,1003,1006,1005]
       }
Student2 = pd.DataFrame(student_data2)

course_data = {
       'Course_Id':[1001,1002,1003,1009],
        'Course_Name':['Statistics','Machine Learning','python','numpy']
       }

Course = pd.DataFrame(course_data)
```

```python
Student1
```

```python
Student2
```

```python
Course
```

### Concatenate Data Frames

```python
all_students = pd.concat([Student1,Student2])
all_students
```

## Merge Data Frames

#### Default Merge


```python
pd.merge(all_students,Course,on='Course_Id')
```

#### right

```python
pd.merge(all_students,Course,on='Course_Id', how='right')
```

#### left

```python
pd.merge(all_students,Course,on='Course_Id',how='left')
```

#### inner

```python
pd.merge(all_students,Course,on='Course_Id',how='inner')
```

#### outer

```python
pd.merge(all_students,Course,on='Course_Id',how='outer')
```
